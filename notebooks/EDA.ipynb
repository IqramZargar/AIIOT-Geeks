# Titanic Dataset - Exploratory Data Analysis (EDA)

## Objective
The goal of this project is to analyze the Titanic dataset and understand patterns that influenced passenger survival.

EDA helps in:
- Understanding dataset structure
- Detecting missing values
- Finding relationships between features
- Extracting meaningful insights


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

## Load Dataset
We load the Titanic dataset using pandas.


```python

df = pd.read_csv("../dataset/train.csv")
df.head()
```

## Dataset Overview

We explore dataset shape, columns, and data types.


```python

df.shape

```

```python
#Column Info:
df.info()
```

## Statistical Summary

This gives statistical details for numerical columns.


```python
df.describe()
```

## Column Description

- Survived: Survival status (0 = No, 1 = Yes)
- Pclass: Passenger class
- Sex: Gender
- Age: Age of passenger
- SibSp: Number of siblings/spouse aboard
- Parch: Number of parents/children aboard
- Fare: Ticket fare
- Embarked: Port of embarkation


## Missing Values Analysis

We check for missing or null values in dataset.


```python
df.isnull().sum()
```

```python
#Missing Values Visualization
sns.heatmap(df.isnull(), cbar=False)
plt.show()
```

### Observation
Age and Cabin contain missing values.
Cabin has a large number of missing entries.


## Univariate Analysis
We analyze distribution of individual columns.


```python
sns.countplot(x="Survived", data=df)
plt.show()
```

Most passengers did not survive compared to those who survived.


```python
sns.countplot(x="Sex", data=df)
plt.show()
```

Most male passengers did not survive compared to females who survived.


```python
sns.histplot(df["Age"], kde=True)
plt.show()
```

Most passengers belong to age group 20-40

## Bivariate Analysis
We analyze relationship between two variables.


```python
sns.countplot(x="Survived", hue="Sex", data=df)
plt.show()
```

Females had higher survival rate compared to males.


```python
sns.countplot(x="Survived", hue="Pclass", data=df)
plt.show()
```

Lower Passenger class determines higher death rate
While survival rate was similar among Three classes


```python
sns.boxplot(x="Survived", y="Age", data=df)
plt.show()
```

Survivors tend to be slightly younger, with more children surviving compared to non-survivors.
However, adult age distributions overlap heavily, so age alone isnt a strong survival predictor.

## Correlation Analysis
Shows relationship between numerical columns.


```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```

## Final Insights

- Female passengers had higher survival chances.
- Higher passenger class had better survival.
- Age distribution shows most passengers were adults.
- Missing values exist in Age and Cabin columns.
